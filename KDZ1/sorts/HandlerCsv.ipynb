```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("100-4100.4_op.csv", sep=';')
```

```python
df
```

```python
df_transposed = df.T
df_transposed
```

```python
plt.plot(df_transposed)
plt.xlabel('кол-во элементов в массиве')
plt.ylabel('время, нс')
plt.title('Сортировка Шелла (последовательность Циура)')
plt.legend(['0-5', '0-4000', 'Почти отсортированный', '4100-1'])
plt.show()
```

```python
plt.plot(df_transposed)
plt.xlabel('кол-во элементов в массиве')
plt.ylabel('кол-во элементарных операций')
plt.title('Сортировка Шелла (последовательность Циура)')
plt.legend(['0-5', '0-4000', 'Почти отсортированный', '4100-1'])
plt.show()
```
