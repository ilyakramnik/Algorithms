```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("dnk_100k_time_4_symb.csv", sep=';')
```

```python
df.pop('Unnamed: 6')
df
```

```python
df_transposed = df.T
df_transposed
```

```python
plt.plot(df_transposed)
plt.xlabel('кол-во элементов в шаблоне')
plt.ylabel('время, нс')
plt.title('10000 символов, бинарный алфавит')
plt.legend(['Наивный алгоритм', 'КМП (стандартные грани)', 'КМП (уточненные грани)'])
plt.show()
```
