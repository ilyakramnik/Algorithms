```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv("dijkstra_ca.csv", sep=';')
df2 = pd.read_csv("floydWarshall_ca.csv", sep=';')
df3 = pd.read_csv("bellmanFord_ca.csv", sep=';')
df4 = pd.read_csv("astar_ca.csv", sep=';')
```

```python
merged_df = pd.concat([df1, df2, df3, df4], axis=0)
merged_df
```

```python
df_transposed1 = df1.T
df_transposed2 = df2.T
df_transposed3 = df3.T
df_transposed4 = df4.T
merged_df_transposed = merged_df.T
```

```python
plt.plot(df_transposed1)
plt.xlabel('кол-во вершин')
plt.ylabel('время, нс')
plt.title('Алгоритм Дейкстры, разреженный граф')
plt.show()
```

```python
plt.plot(df_transposed2)
plt.xlabel('кол-во вершин')
plt.ylabel('время, нс')
plt.title('Алгоритм Флойда-Уоршела, разреженный граф')
plt.show()
```

```python
plt.plot(df_transposed3)
plt.xlabel('кол-во вершин')
plt.ylabel('время, нс')
plt.title('Алгоритм Форда-Беллмана, разреженный граф')
plt.show()
```

```python
plt.plot(df_transposed4)
plt.xlabel('кол-во вершин')
plt.ylabel('время, нс')
plt.title('Алгоритм А*, разреженный граф')
plt.show()
```

```python
plt.plot(merged_df_transposed)
plt.xlabel('кол-во вершин')
plt.ylabel('время, нс')
plt.title('Все алгоритмы, разреженный граф')
plt.legend(['Алгоритм Дейксты', 'Алгоритм Флойда-Уоршела',
           'Алгоритм Форда-Беллмана', 'Алгоритм А*'])
plt.show()
```
